### Lecture 7 - Solving partial differential equations

- using Chebychev polynomials to represent a function

$ f(x) = \sum^{n}_{i=0} c^{i} T_i(x) $

$ <T_j|f> = \int T_j(x) f(x) \frac{1}{\sqrt{1-x^2}} = \sum^{n}_{i=0} c^i \int T_j(x) T_i(x) \frac{1}{\sqrt{1-x^2}} $

$c_j = \kappa <f(x)|T_j>$ where $\kappa$ is some constant. this is roughly correct. jist is that you take some kind of inner product to work out coefficients that allow you to represent your function f(x) in the basis of Chebychev polynomials
